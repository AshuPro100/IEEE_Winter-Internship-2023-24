```python
y = [149,150,225,89,80,200,60,79]
names = ["John", "Jennifer", "Elisabeth", "Lisa", "Laura", "Chris", "Garon", "Shunichi"]
#Bar graph showing name x prices

plt.bar(names, y, width=0.6, color='skyblue')  # Adjust width and color as needed

# Set x-axis ticks and labels
plt.xticks(range(len(names)), names, rotation=90)  # Rotate labels for readability

plt.xlabel("Name")
plt.ylabel("Prices")
plt.title("Prices paid")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
y = [149,150,225,89,80,200]
names = ["John", "Jennifer", "Elisabeth", "LisaRoxanne", "Laura", "Chris"]
#Pie chart showing prices paid by customers

total = sum(y)
y_percentages = [value / total * 100 for value in y]

# Create pie chart
plt.pie(y_percentages, labels=names, autopct="%1.1f%%", startangle=180)  # Customize labels, percentages, etc.

plt.title("Prices paid by the customers")

plt.show()
```

```python
#showing nights spent by the customers by using step graph

import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Names for the x-axis
names = ["John", "Jennifer", "Elisabeth", "Lisa", "Laura", "Chris"]
s=len(names)+1
# Data for the y-axis (assuming you want to keep the same data)
y = [1,1,3,1,10,3]

# Plot with new x-axis values
fig, ax = plt.subplots()
ax.stem(names, y)

# Set x-axis tick positions and labels
ax.set_xticks(np.arange(len(names)))  # Set ticks at each name
ax.set_xticklabels(names, rotation=70)  # Rotate labels for readability

# Adjust limits if needed
ax.set(xlim=(-0.5, s -1.5), ylim=(0, 15))  # Adjust y-axis limit if needed

plt.xlabel("Name")
plt.ylabel("Nights")
plt.title("Graph of Nights Spent by ID")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
